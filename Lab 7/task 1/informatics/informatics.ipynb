# While Loop

## A

```python
k = 1
n = int(input())
while k ** 2 <= n:
    print(k**2)
    k += 1
```

```python
k = 2
n = int(input())
while n % k != 0:
    k += 1

print(k)
```

## C

## D

```python
n = int(input())

while n % 2 == 0 and n > 1:
    n //= 2

if n == 1:
    print("YES")
else:
    print("NO")
```

## E

```python
n = int(input())
count = 0

while (1 << count) < n:
    count += 1

print(count)
```

## E

```python
n = int(input())
arr = list(map(int, input().split()))
happen = False

for i in range(1, len(arr)):
    if (arr[i] > 0 and arr[i - 1] > 0) or (arr[i] < 0 and arr[i - 1] < 0):
        happen = True
        break

if happen == True:
    print("YES")
else:
    print("NO")
```

## F

```python
n = int(input())
arr = list(map(int, input().split()))
count = 0

for i in range(1, len(arr) - 1):
    if arr[i] > arr[i - 1] and arr[i] > arr[i + 1]:
        count += 1

print(count)
```

## G

```python
n = int(input())
arr = list(map(int, input().split()))

for i in range(n // 2):
    arr[i], arr[n - 1 - i] = arr[n - 1 - i], arr[i]

print(*arr)
```

# Functions

## A

```python
def min4(a, b, c, d):
    return min(a, b, c, d)

a, b, c, d = map(int, input().split())

print(min4(a, b, c, d))
```

## B

```python
def power(a, n):
    return a ** n

a, n = map(int, input().split())

print(power(a, n))
```

## C

```python
def xor(x, y):
    if (x == 1 and y == 0) or (x == 0 and y == 1):
        return 1
    else:
        return 0
    
x, y = map(int, input().split())
print(xor(x, y))
```
